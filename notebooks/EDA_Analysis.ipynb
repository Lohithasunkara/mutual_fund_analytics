```python
import pandas as pd
import numpy as np
import os

os.makedirs("../data", exist_ok=True)

# ── 1. nav_data.csv ──────────────────────────────────────────
dates = pd.date_range("2022-01-01", "2026-01-01", freq="B")
schemes = [f"Scheme_{i:02d}" for i in range(1, 41)]
nav_rows = []
for s in schemes:
    nav = 100.0
    for d in dates:
        nav *= (1 + np.random.normal(0.0003, 0.008))
        nav_rows.append({"Date": d, "Scheme_Name": s, "NAV": round(nav, 4)})
pd.DataFrame(nav_rows).to_csv("../data/nav_data.csv", index=False)
print(" nav_data.csv")

# ── 2. aum_data.csv ──────────────────────────────────────────
fund_houses = ["SBI", "HDFC", "ICICI", "Axis", "Kotak",
               "Nippon", "Mirae", "UTI", "Franklin", "DSP"]
aum_rows = []
base = {"SBI":1250000,"HDFC":900000,"ICICI":850000,"Axis":600000,
        "Kotak":550000,"Nippon":400000,"Mirae":350000,
        "UTI":300000,"Franklin":250000,"DSP":200000}
for year in [2022, 2023, 2024, 2025]:
    for fh in fund_houses:
        growth = 1 + (year - 2022) * 0.12
        aum_rows.append({"Fund_House": fh, "Year": year,
                         "AUM_Cr": round(base[fh] * growth)})
pd.DataFrame(aum_rows).to_csv("../data/aum_data.csv", index=False)
print(" aum_data.csv")

# ── 3. sip_inflows.csv ───────────────────────────────────────
months = pd.date_range("2022-01-01", "2025-12-01", freq="MS")
sip_start = 11000
sip_rows = []
for i, m in enumerate(months):
    inflow = sip_start + i * 420 + np.random.randint(-200, 200)
    if m == pd.Timestamp("2025-12-01"):
        inflow = 31002          # all-time high
    sip_rows.append({"Month": m.strftime("%Y-%m-%d"), "SIP_Inflow_Cr": inflow})
pd.DataFrame(sip_rows).to_csv("../data/sip_inflows.csv", index=False)
print(" sip_inflows.csv")

# ── 4. category_inflows.csv ──────────────────────────────────
categories = ["Large Cap","Mid Cap","Small Cap","ELSS",
              "Debt","Hybrid","Index","Sectoral"]
cat_rows = []
for m in pd.date_range("2022-01-01", "2025-12-01", freq="MS"):
    for cat in categories:
        cat_rows.append({"Month": m.strftime("%b-%Y"), "Category": cat,
                         "Net_Inflow_Cr": round(np.random.uniform(-2000, 8000), 1)})
pd.DataFrame(cat_rows).to_csv("../data/category_inflows.csv", index=False)
print(" category_inflows.csv")

# ── 5. investor_demographics.csv ─────────────────────────────
n = 2000
age_groups = np.random.choice(["18-25","26-35","36-45","46-55","55+"],
                               n, p=[0.15,0.35,0.28,0.14,0.08])
genders = np.random.choice(["Male","Female","Other"], n, p=[0.58,0.40,0.02])
sip_amt = [np.random.randint(500,5000) if a=="18-25" else
           np.random.randint(2000,15000) if a=="26-35" else
           np.random.randint(5000,25000) for a in age_groups]
pd.DataFrame({"Age_Group": age_groups,
              "Gender": genders,
              "SIP_Amount": sip_amt[:n]}).to_csv(
    "../data/investor_demographics.csv", index=False)
print(" investor_demographics.csv")

# ── 6. geographic_data.csv ───────────────────────────────────
states = ["Maharashtra","Karnataka","Delhi","Tamil Nadu","Gujarat",
          "Telangana","West Bengal","Rajasthan","UP","Pune"]
tiers  = ["T30","T30","T30","T30","T30","B30","B30","B30","B30","T30"]
geo_rows = [{"State": s, "SIP_Amount": np.random.randint(50000,500000),
             "City_Tier": t} for s, t in zip(states, tiers)]
pd.DataFrame(geo_rows).to_csv("../data/geographic_data.csv", index=False)
print(" geographic_data.csv")

# ── 7. folio_count.csv ───────────────────────────────────────
folio_months = pd.date_range("2022-01-01", "2025-12-01", freq="MS")
folio_rows = []
for i, m in enumerate(folio_months):
    count = round(13.26 + i * (26.12 - 13.26) / (len(folio_months) - 1), 2)
    milestone = None
    if m == pd.Timestamp("2023-06-01"): milestone = "15 Cr Folios"
    if m == pd.Timestamp("2024-06-01"): milestone = "20 Cr Folios"
    if m == pd.Timestamp("2025-12-01"): milestone = "26.12 Cr Folios"
    folio_rows.append({"Month": m.strftime("%Y-%m-%d"),
                       "Folio_Count_Cr": count, "Milestone": milestone})
pd.DataFrame(folio_rows).to_csv("../data/folio_count.csv", index=False)
print(" folio_count.csv")

# ── 8. portfolio_holdings.csv ────────────────────────────────
sectors = ["Banking","IT","Pharma","Auto","FMCG",
           "Energy","Infra","Metals","Telecom","Realty"]
hold_rows = []
for s in schemes:
    weights = np.random.dirichlet(np.ones(len(sectors))) * 100
    for sec, w in zip(sectors, weights):
        hold_rows.append({"Scheme_Name": s, "Sector": sec,
                          "Weight_Pct": round(w, 2)})
pd.DataFrame(hold_rows).to_csv("../data/portfolio_holdings.csv", index=False)
print(" portfolio_holdings.csv")

print("\n All 8 CSV files created in ../data/")
```
